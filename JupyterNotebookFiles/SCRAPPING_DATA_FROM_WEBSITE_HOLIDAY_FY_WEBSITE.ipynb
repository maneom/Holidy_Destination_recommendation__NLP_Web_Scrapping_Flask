```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# REQUEST TO WEBSITE
r = requests.get("https://www.holidify.com/country/india/places-to-visit.html?pageNum=0")
print("STATUS CODE OF WEBSITE = ",r.status_code)
```

```python
# HTML CONTENT OF WEBSITE'S PAGE 1
bsoup = BeautifulSoup(r.text, 'lxml')
bsoup
```

# SCRAPPING DATA

```python
# CREATING EMPTY DATAFRAME TO STOE DATA IN IT
data = pd.DataFrame(columns = ['city', 'url', 'desc'])
```

__THERE ARE TATAL 3 DIFFERENT WEB PAGES WHERRE REQUIRED INFORMATION IS SHOCASED__

I have used for loop to visit each page and scrape required data

```python
for i in range(3):
    r = requests.get("https://www.holidify.com/country/india/places-to-visit.html?pageNum="+str(i))
    bsoup = BeautifulSoup(r.text, 'lxml')
    url = []
    places = []
    desc = []
    # URL
    place = bsoup.find_all("div", class_="position-relative")
    url = [i.find('img').attrs['data-original'].replace(' ','') for i in place]
    # Name
    places = [i.string.split()[1] for i in bsoup.find_all("h3", class_="card-heading")]
    # Description
    desc = [i.string for i in bsoup.find_all('p', class_="card-text")]
    data = pd.concat((data, pd.DataFrame({'city':places, 'url':url, 'desc':desc })))
```

# STORING DATA TO CSV FILE

```python
data.to_csv("data.csv",index=False)
```
